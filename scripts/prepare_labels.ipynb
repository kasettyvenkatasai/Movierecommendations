# LOADING DATASET

```python
import os
import pandas as pd

csv_path = os.path.join("..", "data", "Top_10000_Movies_IMDb.csv")
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    print(df)
else:
    print(f"File not found: {csv_path}")
```

```python
# Multi-label binarizer
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df["Genre"])

# Save encoder
joblib.dump(mlb, "../data/genre_mlb.pkl")

# Save prepared dataset
df.to_csv("../data/movies_prepared.csv", index=False)

print("✅ Labels prepared and saved. Classes:", mlb.classes_)
```
